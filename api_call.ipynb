```python
%load_ext autoreload
%autoreload 2
import os,sys
import time
import requests
from requests.auth import HTTPBasicAuth
sys.path.append('./')
from visualize import vis
```

```python
def get_response(im_url,user,password):
    url='http://13.250.109.108/recognize'
    auth=HTTPBasicAuth(user, password)
    data={'im_url': '{}'.format(im_url)}
    start=time.time()
    response = requests.post(url, json=data,auth=auth)
    print ('total delay = %.3fs'%(time.time()-start))
    print(response)
    return response
```

```python
user = '***'
password = '***'
im_url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/d/d9/Nepali_woman%2C_Ghyaru_%28crop%29.jpg/1200px-Nepali_woman%2C_Ghyaru_%28crop%29.jpg'
response = get_response(im_url,user,password)
vis(im_url,response)
```
